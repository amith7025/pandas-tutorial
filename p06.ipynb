```python
import math
```

```python
def Ncr(n,k):
    f = math.factorial
    return f(n)/(f(k)*f(n-k))
    
```

```python
a = Ncr(5,2)
a
```

```python
def birthday_problem(n):
    value = (1-((364/365)**Ncr(n,2)))
    return value
```

```python
b = birthday_problem(10)
b
```
